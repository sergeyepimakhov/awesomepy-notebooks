# LOCATE

Command that searches for files by name on the entire system.

## SYNOPSIS

```bash
locate [option(s)] pattern ...
```

## EXAMPLES

### Find a file by name

````python
%%bash
locate pwd.h
````

### Show only first N matches

````python
%%bash 
locate -l 1 pwd.h
````

### Find a file by extension

````python
%%bash
locate *.jpg
````

### Find a file by pattern

````python
%%bash
locate -l 3 passwd
````

### Count the number of matches

````python
%%bash
locate -c pwd.h
````

### Case insensitive search

````python
%%bash
locate -i -l 3 *.jpg
````
